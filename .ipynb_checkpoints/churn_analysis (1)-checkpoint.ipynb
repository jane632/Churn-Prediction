# **1.Business Understanding**


## 1.1 Background Information

#### Customer churn happens when people stop using a company’s services.In the telecommunications industry, this can occur if customers aren't happy with the service or find better deals elsewhere. Churn is a big issue  because it means losing customers, which can lead to loss in  business.
#### Churn management is about finding customers who might leave and find ways to keep them. This is done by looking at customer data to understand their behavior and predict if they’re at risk of leaving. Once they’re spotted, the company can offer better customer service to keep them.

#### To address customer churn at SyriaTel, a machine learning model will be built to analyze customer data to predict which customers are at risk of leaving. First, an "exploratory data analysis" will be conducted to understand its structure. Then, important features will be extracted to feed into the model. Various algorithms, like logistic regression, will be used to train the model. Its performance will be evaluated, and adjustments will be made to handle issues like data imbalance. This approach will help SyriaTel to identify at-risk customers and  improve services to retain them.


#### A predictive model for customer churn can lead to improvements in customer retention and the performance of the business. This enables the company, "SyriaTel" to reduce the loss in revenue and understanding of why causes customers want to leave. This, in turn, enables the company to remain competitive in the market due to increased profitability,increased customer satisfaction and enhanced retention rates.

* ## Checking for Missing Values

```python
#checks the sum of missing values
def check_duplicates_and_missing_values(df):
    #checking for missing values
    missing_values = churn_analysis_df.isnull().sum() 
    missing_percentage = (missing_values/ len(df))* 100
    missing_info = pd.DataFrame({'Missing Values': missing_values, 'Percentage': missing_percentage})
    #checking for duplicates
    duplicated_rows = df.duplicated()
    if duplicated_rows.any():
        print('Duplicates found: {}', duplicated_rows)
    else:
        print('No duplicates')

    #return missing values info and duplicate count
    return missing_info , duplicated_rows
    
    

check_duplicates_and_missing_values(churn_analysis_df)
```

### 2. Distribution of categorical features

```python
#Checking the distribution of categorical features
def plot_categorical_features(data, feature):
    """
    Plots the distribution of a categorical feature in the given data.
    """
    plt.figure(figsize=(10, 5))
    sns.countplot(x=feature, data=data, color='pink', order=data[feature].value_counts().index)
    plt.xticks(rotation=90)
    plt.title(f"Distribution of {feature}", fontsize=16) #Adds title
    plt.show()
    
```

## Bivariate Analysis

###  "Let’s analyze how the target values interact with other features in the dataset to uncover useful patterns and insights.

#### 1. Relationship between categorical variables(voice mail plan,international plan and state) and the churn rate

```python
#Checking the distribution of categorical features based on churn rate
def plot_categorical_distribution_on_churn_rate(df, feature):
    """
    Plots the distribution of a categorical feature in the given data.
    """
    plt.figure(figsize=(10, 4))
    churn_counts = df.groupby(feature)["churn"].sum().sort_values(ascending=False)
    top_10_categories = churn_counts.head(10).index.tolist()
    sns.countplot(x=feature, hue="churn", data=df, order=top_10_categories ,palette=["#FF6347", "#4682B4"])
    plt.xticks(rotation=90)
    plt.legend(loc="upper right")
    plt.title(f"Distribution of '{feature}' Based on Churn Rate", fontsize=16)
    plt.show()
```

* a. International plan

```python
plot_categorical_distribution_on_churn_rate(churn_analysis_df2, 'international plan')
```

* Insight 

Most of the customers who churned didnt have an international_plan.

* b. Voice mail plan

```python
plot_categorical_distribution_on_churn_rate(churn_analysis_df2, 'voice mail plan')
```

* Insight

Most of the customers who churned didnt have a voice mail plan

* c. State

```python
 plot_categorical_distribution_on_churn_rate(churn_analysis_df2, 'state')
```
